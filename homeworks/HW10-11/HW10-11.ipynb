### 3.1. Часть A (S10): классификация, аугментации и transfer learning (обязательно)

Задача: показать эффект CNN, аугментаций и transfer learning.

Провести 4 эксперимента (C1-C4). Эксперимент = фиксированный конфиг модели/обучения + обучение + оценка на `val`.

- **C1 (simple-cnn-base)**: простая CNN без аугментаций.
- **C2 (simple-cnn-aug)**: та же CNN, но с разумными аугментациями.
- **C3 (resnet18-head-only)**: `ResNet18` с pretrained weights; backbone заморожен, обучается только классификационная голова.
- **C4 (resnet18-finetune)**: `ResNet18` с pretrained weights; частичное fine-tune (например, `layer4 + fc`, либо другой близкий по смыслу вариант).

Требования:

- loss: `CrossEntropyLoss`;
- основная метрика: accuracy;
- сравнение должно быть корректным: фиксированный split, фиксированный seed, сопоставимые условия обучения;
- лучшая модель части A выбирается по `best_val_accuracy`;
- test-часть используется **один раз**: только для финальной оценки лучшей модели после выбора по `val`.

Важно:

- в C1 и C2 архитектура CNN должна быть одной и той же, меняются только аугментации;
- в C3 и C4 обязательно использовать именно pretrained `ResNet18` из `torchvision.models`;
- в отчёте нужно явно объяснить, что дали аугментации и что дало partial fine-tuning относительно head-only.
