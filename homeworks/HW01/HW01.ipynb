4) Примеры индексирования и срезов

```python
s = np.random.random(5)
s
```

```python
e1 = s[4] #4 индекс элемента
e1
```

```python
e2 = s[1:] #с индекса 1 до конца
e2
```

```python
e3 = s[::-1] #обратный порядок
e3
```

```python
s2 = np.random.random((5,5))
s2
```

```python
e4 = s2[2,3] #элемент в 3 строке и 4 столбце
e5 = s2[3,4] #элемент в 4 строке и 5 столбце
print(e4, e5)
```

5) Примеры векторных операций и базовых статистик

```python
y = np.array([1, 2, 3, 4])
u = np.array([10, 20, 30, 40])
c_mul = u * y      #Поэлементное сложение
c_sum = u+y        #Поэлементное сложение
print("Сумма:", c_sum)
print("Произведение:", c_mul)
```

```python
m1 = np.array([[1, 2],
               [3, 4]])

m2 = np.array([[10, 20],
               [30, 40]])
print(m1 + m2)
print(m1 * m2)
```

```python
m = np.array([[1, 2, 3],
              [4, 5, 6]])

print("Среднее по всем элементам:", m.mean()) # Среднее по всей матрице
print("Среднее по строкам:", m.mean(axis=1))  # Среднее по строкам (axis=1)
print("Среднее по столбцам:", m.mean(axis=0)) # Среднее по столбцам (axis=0)
```

```python
t = np.random.sample(5)
print("Матрица:", t)
print("Среднее:", t.mean())                 # Среднее значение
print("Сумма:", t.sum())                    # Сумма
print("Минимум:", t.min())                  # Минимум
print("Максимум:", t.max())                 # Максимум
```
